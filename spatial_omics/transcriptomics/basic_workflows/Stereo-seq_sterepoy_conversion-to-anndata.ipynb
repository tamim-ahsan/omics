# 1. Import packages

```python
import stereo as st
print(f"steropy=={st.__version__}")
```

# 2. Read data 
The same data as this one (https://github.com/tamim-ahsan/omics/blob/main/spatial_omics/transcriptomics/basic_workflows/Stereo-Seq_stereopy_square-bin-workflow.ipynb) was used

```python
file_path = "GSM8330060_B02009F6.tissue.gef"
data = st.io.read_gef(file_path = file_path,
                     bin_size = 50)
data
```

# 3. Save the file as an AnnData object

```python
adata = st.io.stereo_to_anndata(data,
                               flavor = "scanpy",
                               output = "GSM8330060_B02009F6_square.h5ad")
```
